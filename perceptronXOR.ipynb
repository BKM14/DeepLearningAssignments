```python
import numpy as np
```

```python
class Perceptron:
    
    def __init__(self, wNOT, wAND, wOR, learning_rate):
        self.wNOT = wNOT
        self.wAND = wAND
        self.wOR = wOR
        self.learning_rate = learning_rate
    
    def unitStep(self, v):
        if v >= 0:return 1
        else: return 0
    
    def perceptronModel(self, x, w, b):
        v = np.dot(w, x) + b
        y = self.unitStep(v)
        return y
    
    def NOT_logicFunction(self, x):
        bNOT = 0.5
        result = self.perceptronModel(x, self.wNOT, bNOT)
        return result
        
    
    def AND_logicFunction(self, x):
        bAND = -1.5
        result = self.perceptronModel(x, self.wAND, bAND)
        return result
    
    def OR_logicFunction(self, x):
        bOR = -0.5
        result = self.perceptronModel(x, self.wOR, bOR)
        return result
    
    def XOR_logicFunction(self, x):
        y1 = self.AND_logicFunction(x)
        y2 = self.OR_logicFunction(x)
        y3 = self.NOT_logicFunction(y1)
        final_x = np.array([y2, y3])
        final_output = self.AND_logicFunction(final_x)
        return final_output

    def weight_update(self, expected_value, predicted_value, old_weight, x):
        new_weight = old_weight + (self.learning_rate*(expected_value - predicted_value)*x)
        return new_weight
    
    def update_and(self):
        input_value = np.array([(0, 0), (0, 1), (1, 0), (1, 1)])
        expected_value = np.array([0, 0, 0, 1])
        
        i = 0
        while (i < len(input_value)):
            output_value = self.AND_logicFunction(input_value[i])
            if output_value != expected_value[i]:
                self.wAND = self.weight_update(expected_value[i], output_value, self.wAND, input_value[i])
                i = 0
            else: i += 1
            
    def update_or(self):
        input_value = np.array([(0, 0), (0, 1), (1, 0), (1, 1)])
        expected_value = np.array([0, 1, 1, 1])
        
        i = 0
        while (i < len(input_value)):
            output_value = self.OR_logicFunction(input_value[i])
            if output_value != expected_value[i]:
                self.wOR = self.weight_update(expected_value[i], output_value, self.wOR, input_value[i])
                i = 0
            else: i += 1
            
    def update_not(self):
        input_value = np.array([0, 1])
        expected_value = np.array([1, 0])
        
        i = 0
        while (i < len(input_value)):
            output_value = self.NOT_logicFunction(input_value[i])
            if output_value != expected_value[i]:
                self.wNOT = self.weight_update(expected_value[i], output_value, self.wNOT, input_value[i])
                i = 0
            else: i += 1
```

```python
test1 = np.array([0, 1])
test2 = np.array([1, 1])
test3 = np.array([0, 0])
test4 = np.array([1, 0])
output = [1, 0, 0, 1]

test_data = [test1, test2, test3, test4]

perceptron = Perceptron(wNOT = 2, wAND=np.array([0, 1]), wOR=np.array([1, 1]), learning_rate=0.3)
perceptron.update_and()
perceptron.update_or()
perceptron.update_not()

for i in range(len(test_data)):
    result = perceptron.XOR_logicFunction(test_data[i])
    print(result)
    
print(perceptron.wOR, perceptron.wAND, perceptron.wNOT)
```
